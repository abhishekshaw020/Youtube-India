```python

plt.figure(figsize=(12, 6))
sns.countplot(x='categoryId', data=df, palette='viridis')
plt.title('Distribution of Videos by Category')
plt.xlabel('Category ID')
plt.ylabel('Number of Videos')
plt.show()
```

```python
# Scatter plot for likes vs. views
plt.figure(figsize=(10, 6))
sns.scatterplot(x='likes', y='view_count', data=df, alpha=0.7)
plt.title('Scatter Plot of Likes vs. Views')
plt.xlabel('Number of Likes')
plt.ylabel('Number of Views')
plt.show()
```

```python
# Bar chart for video comments
top_comments = df.sort_values(by='comment_count', ascending=False).head(25)
plt.figure(figsize=(12, 6))
sns.barplot(x='title', y='comment_count', data=top_comments, palette='muted')
plt.title('Top  Videos with Highest Comment Count')
plt.xlabel('Video Title')
plt.ylabel('Number of Comments')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Boxplot for video ratings
plt.figure(figsize=(8, 6))
sns.boxplot(x='ratings_disabled', y='likes', data=df)
plt.title('Boxplot of Likes with Ratings Disabled/Enabled')
plt.xlabel('Ratings Disabled')
plt.ylabel('Number of Likes')
plt.show()
```

```python
# Pie chart for video comments disabled
comments_disabled_counts = df['comments_disabled'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(comments_disabled_counts, labels=comments_disabled_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Videos by Comments Disabled')
plt.show()
```

```python
# Convert 'publishedAt' to datetime
df['publishedAt'] = pd.to_datetime(df['publishedAt'])

# Extract year and month
df['publish_year'] = df['publishedAt'].dt.year
df['publish_month'] = df['publishedAt'].dt.month

# Line chart for views over time
plt.figure(figsize=(12, 6))
views_over_time = df.groupby(['publish_year', 'publish_month'])['view_count'].sum()
views_over_time.plot(marker='o')
plt.title('Total Views Over Time')
plt.xlabel('Year, Month')
plt.ylabel('Total Views')
plt.grid(True)
plt.show()
```

```python
# Bar chart for top channels by views
top_channels = df.groupby('channelTitle')['view_count'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_channels.values, y=top_channels.index, palette='muted')
plt.title('Top 10 Channels by Total Views')
plt.xlabel('Total Views')
plt.ylabel('Channel Title')
plt.show()
```

```python
# Stacked bar chart for likes and dislikes
likes_dislikes_data = df[['likes', 'dislikes']].head(20)
likes_dislikes_data.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Likes and Dislikes Comparison for Top 20 Videos')
plt.xlabel('Video Index')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot for comments by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='categoryId', y='comment_count', data=df)
plt.title('Boxplot of Comments by Category')
plt.xlabel('Category ID')
plt.ylabel('Number of Comments')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Heatmap for correlation between variables
correlation_matrix = df[['view_count', 'likes', 'dislikes', 'comment_count']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Violin plot for distribution of likes by category
plt.figure(figsize=(12, 6))
sns.violinplot(x='categoryId', y='likes', data=df, palette='viridis')
plt.title('Distribution of Likes by Category')
plt.xlabel('Category ID')
plt.ylabel('Number of Likes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Boxplot for views and likes with comments disabled/enabled
plt.figure(figsize=(10, 6))
sns.boxplot(x='comments_disabled', y='view_count', hue='ratings_disabled', data=df)
plt.title('Boxplot of Views with Comments Disabled/Enabled and Ratings Disabled/Enabled')
plt.xlabel('Comments Disabled')
plt.ylabel('Number of Views')
plt.show()
```

```python
# Bar chart for videos with disabled comments and ratings
disabled_data = df.groupby(['comments_disabled', 'ratings_disabled']).size().reset_index(name='count')
plt.figure(figsize=(8, 6))
sns.barplot(x='comments_disabled', y='count', hue='ratings_disabled', data=disabled_data, palette='coolwarm')
plt.title('Videos with Disabled Comments and Ratings')
plt.xlabel('Comments Disabled')
plt.ylabel('Number of Videos')
plt.show()
```

```python
# Cumulative views over time
df.sort_values('publishedAt', inplace=True)
df['cumulative_views'] = df['view_count'].cumsum()

plt.figure(figsize=(12, 6))
plt.plot(df['publishedAt'], df['cumulative_views'], marker='o')
plt.title('Cumulative Views Over Time')
plt.xlabel('Date Published')
plt.ylabel('Cumulative Views')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Donut chart for likes vs. dislikes proportion
likes_sum = df['likes'].sum()
dislikes_sum = df['dislikes'].sum()

plt.figure(figsize=(8, 8))
plt.pie([likes_sum, dislikes_sum], labels=['Likes', 'Dislikes'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Likes vs. Dislikes Proportion')
plt.show()
```

```python
# Area chart for cumulative likes and dislikes
df.sort_values('publishedAt', inplace=True)
df['cumulative_likes'] = df['likes'].cumsum()
df['cumulative_dislikes'] = df['dislikes'].cumsum()

plt.figure(figsize=(12, 6))
plt.fill_between(df['publishedAt'], df['cumulative_likes'], label='Cumulative Likes', alpha=0.7)
plt.fill_between(df['publishedAt'], df['cumulative_dislikes'], label='Cumulative Dislikes', alpha=0.7)
plt.title('Cumulative Likes and Dislikes Over Time')
plt.xlabel('Date Published')
plt.ylabel('Cumulative Count')
plt.legend()
plt.show()
```

```python
# Scatter plot for likes vs. comments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='likes', y='comment_count', data=df, alpha=0.7)
plt.title('Scatter Plot of Likes vs. Comments')
plt.xlabel('Number of Likes')
plt.ylabel('Number of Comments')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter plot for likes, dislikes, and views
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['likes'], df['dislikes'], df['view_count'], c='teal', marker='o', alpha=0.5)

ax.set_xlabel('Likes')
ax.set_ylabel('Dislikes')
ax.set_zlabel('Views')
ax.set_title('3D Scatter Plot for Likes, Dislikes, and Views')
plt.show()
```
